```python
import pandas as pd
```

```python
fname = "steam-project\data\steam_store_details_2026-03-18.csv"
steam_api = pd.read_csv(fname)
steam_api.head()
```

```python
steam_api.info()
steam_api.describe()    
```

```python
# Adding ingestion date column
import re
m = re.search(r"(\d{4}-\d{2}-\d{2})", fname)
if m:
    steam_api["ingestion_date"] = pd.to_datetime(m.group(1))

steam_api.head()
```

```python
# Droping metacritic_score column - too many missing values
steam_api.drop(columns=["metacritic_score"], inplace=True)

# Droping series column - too many missing values
steam_api.drop(columns=["series"], inplace=True)

steam_api.head()
```

```python
# Converting release_date to datetime, coercing errors to NaT
steam_api["release_date"] = pd.to_datetime(steam_api["release_date"], errors="coerce")

# Adding a new column to indicate if release_date is missing
steam_api["release_date_missing"] = steam_api["release_date"].isna()

steam_api.head()
```

```python
# Removing currency symbols from price columns and converting to numeric, filling missing values with 0 (free-to-play games)
steam_api['price_usd'] = steam_api['price_usd'].astype(str).str.replace(r'[^\d\.\-]', '', regex=True)
steam_api['price_pln'] = (
    steam_api['price_pln'].astype(str)
    .str.replace(r'[^\d,\-\s\(\)]', '', regex=True)  # Remove non-numeric characters except for digits, comma, minus, parentheses, and whitespace
    .str.replace(r'\s+', '', regex=True)  # Remove any remaining whitespace
    .str.replace(',', '.', regex=True) # Replace comma with dot for decimal conversion
    .pipe(pd.to_numeric, errors='coerce')
)
steam_api["price_usd"] = pd.to_numeric(steam_api["price_usd"], errors="coerce").fillna(0)
steam_api["price_pln"] = pd.to_numeric(steam_api["price_pln"], errors="coerce").fillna(0)

steam_api.head(5)
```

```python
# Create normalized tables

# 1. Games table - all columns except genres and languages
games = steam_api.drop(columns=['genre', 'languages']).copy()
print("Games table created:")
print(f"Shape: {games.shape}")
print(games.head())
```

```python
# 2. Genres table - appid with individual genres (excluding "Free to Play")
genres = steam_api[['appid', 'genre']].copy()
# Split genres by comma and explode into separate rows
genres['genre'] = genres['genre'].str.split(',')
genres = genres.explode('genre')
# Strip whitespace
genres['genre'] = genres['genre'].str.strip()
# Remove "Free To Play" entries
genres = genres[genres['genre'] != 'Free To Play'].reset_index(drop=True)
print("Genres table created:")
print(f"Shape: {genres.shape}")
print(genres.head(10))
```

```python
# 3. Languages table - appid with individual languages and audio support
languages = steam_api[['appid', 'languages']].copy()
# Split languages by comma and explode into separate rows
languages['language_raw'] = languages['languages'].str.split(',')
languages = languages.explode('language_raw')
languages = languages.drop(columns=['languages']).reset_index(drop=True)

# Check if language has audio (indicated by <strong>*</strong>)
languages['audio'] = languages['language_raw'].str.contains(r'<strong>\*</strong>', regex=True)

# Clean up language name by removing <strong>*</strong> and whitespace
languages['language'] = (
    languages['language_raw']
    .str.replace(r'<strong>\*</strong>', '', regex=True)
    .str.strip()
)

# Keep only appid, language, and audio columns
languages = languages[['appid', 'language', 'audio']]

print("Languages table created:")
print(f"Shape: {languages.shape}")
print(languages.head(15))
```

```python
# Summary of normalized tables
print("=" * 60)
print("NORMALIZED TABLES SUMMARY")
print("=" * 60)

print(f"\n1. GAMES TABLE")
print(f"   Shape: {games.shape} (rows, columns)")
print(f"   Columns: {list(games.columns)}")
print(f"\n2. GENRES TABLE")
print(f"   Shape: {genres.shape} (rows, columns)")
print(f"   Columns: {list(genres.columns)}")
print(f"   Unique genres: {genres['genre'].nunique()}")
print(f"   Genres: {sorted(genres['genre'].unique().tolist())}")
print(f"\n3. LANGUAGES TABLE")
print(f"   Shape: {languages.shape} (rows, columns)")
print(f"   Columns: {list(languages.columns)}")
print(f"   Unique languages: {languages['language'].nunique()}")
print(f"   Languages with audio support: {languages['audio'].sum()}")
print(f"   Languages without audio support: {(~languages['audio']).sum()}")
```
