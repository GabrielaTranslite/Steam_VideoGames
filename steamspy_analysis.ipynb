```python
import pandas as pd

fname = "steamspy_top100_games_2026-03-18.csv"
steam_spy = pd.read_csv("steam-project/data/" + fname)
steam_spy.head()
```

```python
# Sorting data by ccu in descending order
steam_spy.sort_values(by="ccu", ascending=False, inplace=True)
steam_spy.head(10)
```

```python
# Dropping unnecessary columns
steam_spy.drop(columns=["developer", "publisher", "score_rank", "userscore", "price", "initialprice", "discount"], inplace=True)
steam_spy.head()
```

```python
# Replacing owners column with categories
# Create a mapping for cleaner category labels
owners_mapping = {
    '5,000,000 .. 10,000,000': '5M-10M',
    '10,000,000 .. 20,000,000': '10M-20M',
    '20,000,000 .. 50,000,000': '20M-50M',
    '50,000,000 .. 100,000,000': '50M-100M',
    '100,000,000 .. 200,000,000': '100M-200M'
}

# Add potential categories for lower and higher counts that might come up later
# Lower: '0 .. 5,000,000': '0-5M'
# Higher: '200,000,000 .. ': '200M+'

steam_spy['categories'] = steam_spy['owners'].map(owners_mapping)
steam_spy.drop(columns=['owners'], inplace=True)
steam_spy.head()
```

```python
# Check for any missing fields
steam_spy.isna().sum()
```

```python
# Check column data types
steam_spy.dtypes
```

```python
# Converting name to string type
steam_spy['name'] = steam_spy['name'].astype('string')
steam_spy.dtypes
# Converting categories to categorical type
steam_spy['categories'] = steam_spy['categories'].astype('category')
steam_spy.dtypes
```

```python
# Adding ingestion date column based on filename
import re
m = re.search(r"(\d{4}-\d{2}-\d{2})", fname)
if m:
    steam_spy["ingestion_date"] = pd.to_datetime(m.group(1))

steam_spy.head()
```
